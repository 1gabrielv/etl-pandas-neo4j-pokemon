```python
print("Iniciando a carga de dados otimizada em lote no Neo4j...")

lista_pokemons = []
for _, linha in df.iterrows():
    lista_pokemons.append({
        'name': str(linha['name']),
        'index': int(linha['index']),
        'hp': int(linha['hp']),
        'attack': int(linha['attack']),
        'defense': int(linha['defense']),
        'sp_atk': int(linha['sp_atk']),
        'sp_def': int(linha['sp_def']),
        'speed': int(linha['speed']),
        'generation': str(linha['generation']),
        'type1': str(linha['type1']),
        'type2': str(linha['type2']),
        'ability1': str(linha['ability1']),
        'ability2': str(linha['ability2']),
        'hidden_ability': str(linha['hidden_ability']),
        'smogon_2022': float(linha['Smogon_VGC_Usage_2022']),
        'smogon_2023': float(linha['Smogon_VGC_Usage_2023']),
        'smogon_2024': float(linha['Smogon_VGC_Usage_2024']),
        'worlds_2022': float(linha['Worlds_VGC_Usage_2022']),
        'worlds_2023': float(linha['Worlds_VGC_Usage_2023']),
        'worlds_2024': float(linha['Worlds_VGC_Usage_2024'])
    })

query_lote = """
UNWIND $linhas AS linha

MERGE (p:Pokemon {name: linha.name})
SET p.pokedex_number = linha.index,
    p.hp = linha.hp,
    p.attack = linha.attack,
    p.defense = linha.defense,
    p.sp_atk = linha.sp_atk,
    p.sp_def = linha.sp_def,
    p.speed = linha.speed,
    p.generation = linha.generation

MERGE (t1:Tipo {nome: linha.type1})    
MERGE (p)-[:POSSUI_TIPO {ordem: "principal"}]->(t1)

FOREACH (_ IN CASE WHEN linha.type2 <> 'None' THEN [1] ELSE [] END |
    MERGE (t2:Tipo {nome: linha.type2})
    MERGE (p)-[:POSSUI_TIPO {ordem: "secundario"}]->(t2)
)

MERGE (a1:Habilidade {nome: linha.ability1})
MERGE (p)-[:POSSUI_HABILIDADE {tipo: "comum"}]->(a1)

FOREACH (_ IN CASE WHEN linha.ability2 <> 'None' THEN [1] ELSE [] END |
    MERGE (a2:Habilidade {nome: linha.ability2})
    MERGE (p)-[:POSSUI_HABILIDADE {tipo: "comum"}]->(a2)
)

FOREACH (_ IN CASE WHEN linha.hidden_ability <> 'None' THEN [1] ELSE [] END |
    MERGE (ha:Habilidade {nome: linha.hidden_ability})
    MERGE (p)-[:POSSUI_HABILIDADE {tipo: "oculta"}]->(ha)
)

MERGE (st22:Torneio {nome: "Smogon VGC", ano: 2022})
MERGE (p)-[r1:COMPETIU_EM]->(st22) SET r1.taxa_uso = linha.smogon_2022

MERGE (st23:Torneio {nome: "Smogon VGC", ano: 2023})
MERGE (p)-[r2:COMPETIU_EM]->(st23) SET r2.taxa_uso = linha.smogon_2023

MERGE (st24:Torneio {nome: "Smogon VGC", ano: 2024})
MERGE (p)-[r3:COMPETIU_EM]->(st24) SET r3.taxa_uso = linha.smogon_2024   

MERGE (wt22:Torneio {nome: "Worlds VGC", ano: 2022})
MERGE (p)-[r4:COMPETIU_EM]->(wt22) SET r4.taxa_uso = linha.worlds_2022

MERGE (wt23:Torneio {nome: "Worlds VGC", ano: 2023})
MERGE (p)-[r5:COMPETIU_EM]->(wt23) SET r5.taxa_uso = linha.worlds_2023   

MERGE (wt24:Torneio {nome: "Worlds VGC", ano: 2024})
MERGE (p)-[r6:COMPETIU_EM]->(wt24) SET r6.taxa_uso = linha.worlds_2024
"""

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        session.run(query_lote, linhas=lista_pokemons)

print("✓ Sucesso! Todos os dados injetados de uma só vez em lote.")
```


## Categoria 1 — Consultas Básicas (Mapeamento de Estrutura Básica)

Top 10 Pokémon com Maior Velocidade Base

```python
query_1 = """
MATCH (p:Pokemon)
RETURN p.name AS Pokemon, p.speed AS Velocidade, p.generation AS Geracao
ORDER BY Velocidade DESC LIMIT 10
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_1).data()))
```

Distribuição de Pokémon por Tipo Primário

```python
query_2 = """
MATCH (p:Pokemon)-[:POSSUI_TIPO {ordem: "principal"}]->(t:Tipo)
RETURN t.nome AS Tipo, COUNT(p) AS QtdPokemon
ORDER BY QtdPokemon DESC
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_2).data()))
```

As 10 Habilidades mais Comuns no Universo Pokémon

## Categoria 2 — CONSULTAS COMPLEXAS (Interconexão Multidimensional de Caminhos)

O Caminho da Dominância Oculta

```python
query_4 = """
MATCH (h:Habilidade)<-[:POSSUI_HABILIDADE {tipo: "oculta"}]-(p:Pokemon)-[r:COMPETIU_EM]->(t:Torneio {ano: 2024})
WHERE r.taxa_uso > 15.0
RETURN p.name AS Pokemon, 
       h.nome AS HabilidadeOculta, 
       r.taxa_uso AS TaxaUso, 
       t.nome AS Torneio
ORDER BY TaxaUso DESC
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_4).data()))
```

Interseção de Tipos Opostos via Sinergia de Habilidade Compartilhada

```python
query_5 = """
MATCH (t1:Tipo {nome: "Fire"})<-[:POSSUI_TIPO]-(p1:Pokemon)-[:POSSUI_HABILIDADE]->(h:Habilidade)
MATCH (t2:Tipo {nome: "Water"})<-[:POSSUI_TIPO]-(p2:Pokemon)-[:POSSUI_HABILIDADE]->(h)
RETURN h.nome AS HabilidadeCompartilhada, 
       p1.name AS PokemonFogo, 
       p2.name AS PokemonAgua
ORDER BY HabilidadeCompartilhada ASC
LIMIT 10
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_5).data()))
```

A Teia Completa do Metagame

```python
query_6 = """
MATCH (t:Tipo)<-[:POSSUI_TIPO {ordem: "principal"}]-(p:Pokemon)-[:POSSUI_HABILIDADE {tipo: "comum"}]->(h:Habilidade)
MATCH (p)-[r:COMPETIU_EM]->(torneio:Torneio {ano: 2024})
WHERE r.taxa_uso > 25.0
RETURN torneio.nome AS Torneio, 
       p.name AS Pokemon, 
       t.nome AS TipoPrincipal, 
       h.nome AS HabilidadeDominante, 
       r.taxa_uso AS TaxaUso
ORDER BY TaxaUso DESC
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_6).data()))
```

## Categoria 3 — Consultas Sérias   (Análise Estratégica do Metagame e Data Analytics)

Centralização do Meta - Concentração de Pokémon no Top Tier (>30% uso) no Worlds

```python
query_7 = """
MATCH (p:Pokemon)-[r:COMPETIU_EM]->(t:Torneio {nome: "Worlds VGC"})
WHERE r.taxa_uso > 30.0
RETURN t.ano AS Ano, 
       p.name AS Pokemon, 
       r.taxa_uso AS TaxaUso, 
       p.generation AS Geracao
ORDER BY Ano ASC, TaxaUso DESC
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_7).data()))
```

Mapeamento de Hiperconsistência Histórica (Presença nos 6 Torneios Simultâneos)

```python
query_8 = """
MATCH (p:Pokemon)-[r:COMPETIU_EM]->(t:Torneio)
WITH p, 
     COUNT(CASE WHEN r.taxa_uso > 0 THEN 1 END) AS TorneiosAtivos, 
     SUM(r.taxa_uso) AS UsoTotal
WHERE TorneiosAtivos = 6
RETURN p.name AS Pokemon, 
       TorneiosAtivos, 
       ROUND(UsoTotal, 2) AS UsoAcumulado, 
       ROUND(UsoTotal / 6, 2) AS MediaPorTorneio
ORDER BY UsoAcumulado DESC
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_8).data()))
```

Mapeamento de Habilidades Ocultas Compartilhadas no Top Tier de 2024

```python
query_9 = """
MATCH (h:Habilidade)<-[:POSSUI_HABILIDADE {tipo: "oculta"}]-(p:Pokemon)-[r:COMPETIU_EM]->(t:Torneio {ano: 2024})
WHERE r.taxa_uso > 15.0
WITH h.nome AS HabilidadeOculta, 
     p.name AS Pokemon, 
     AVG(r.taxa_uso) AS MediaUso
RETURN HabilidadeOculta, 
       Pokemon, 
       ROUND(MediaUso, 2) AS TaxaUso
ORDER BY HabilidadeOculta ASC, TaxaUso DESC
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_9).data()))
```

Análise de Evolução - Top 5 Pokémon Veteranos com Maior Crescimento (2022 vs 2024)

```python
query_10 = """
MATCH (p:Pokemon)-[r22:COMPETIU_EM]->(t22:Torneio {ano: 2022})
MATCH (p)-[r24:COMPETIU_EM]->(t24:Torneio {ano: 2024})
WHERE r22.taxa_uso > 0.0 AND r24.taxa_uso > 0.0
WITH p.name AS Pokemon, 
     AVG(r22.taxa_uso) AS Media22, 
     AVG(r24.taxa_uso) AS Media24
WITH Pokemon, 
     Media22, 
     Media24, 
     (Media24 - Media22) AS Crescimento
RETURN Pokemon, 
       ROUND(Media22, 2) AS UsoMedio_2022, 
       ROUND(Media24, 2) AS UsoMedio_2024, 
       ROUND(Crescimento, 2) AS GanhoDeTaxa
ORDER BY GanhoDeTaxa DESC
LIMIT 20
"""
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    with driver.session() as session:
        display(pd.DataFrame(session.run(query_10).data()))
```
