```python
import pandas as pd
```

```python
dados_brutos = pd.read_csv('pokemon_sujo_para_etl.csv')
df = dados_brutos.copy()
```

```python
df["name"] = df["name"].str.strip()
df["name"] = df["name"].str.lower()
df["name"] = df["name"].str.capitalize()
print(df["name"].head(10))
```

```python
df["type1"] = df["type1"].str.strip()
df["type1"] = df["type1"].str.lower()
df["type1"] = df["type1"].str.capitalize()
df["type2"] = df["type2"].str.strip()
df["type2"] = df["type2"].str.lower()
df["type2"] = df["type2"].str.capitalize()
print(df["type2"].head(10))
print(df["type1"].head(10))
```

```python
df["ability1"] = df["ability1"].str.strip()
df["ability1"] = df["ability1"].str.lower()
df["ability1"] = df["ability1"].str.capitalize()
df["ability2"] = df["ability2"].str.strip()
df["ability2"] = df["ability2"].str.lower()
df["ability2"] = df["ability2"].str.capitalize()

df["ability2"] = df["ability2"].replace("N/a", "None")
df["ability2"] = df["ability2"].replace("No_ability", "None") 
df["ability2"] = df["ability2"].replace("Nan", "None")
df["ability2"] = df["ability2"].replace("Null", "None")
df["ability2"] = df["ability2"].replace("Sem_habilidade", "None")
df["ability2"] = df["ability2"].replace("No-ability", "None")
df["ability2"] = df["ability2"].replace("", "None")
df["ability2"] = df["ability2"].fillna("None")


print(df["ability2"].head(100))
lista = [df["ability2"].iloc[i] for i in range(300)]
print(lista)
```

```python
df["hidden_ability"] = df["hidden_ability"].str.strip()
df["hidden_ability"] = df["hidden_ability"].str.lower()
df["hidden_ability"] = df["hidden_ability"].str.capitalize()

print(df["hidden_ability"].head(100))
```

```python
df["total_stats"] = df["hp"] + df["attack"] + df["defense"] + df["sp_atk"] + df["sp_def"] + df["speed"]
```

```python
romanos = {'i': '1', 'ii': '2', 'iii': '3', 'iv': '4', 'v': '5', 'vi': '6', 'vii': '7', 'viii': '8', 'ix': '9'}

df["generation"] = df["generation"].astype(str)
df["generation"] = df["generation"].str.strip()
df["generation"] = df["generation"].str.lower()
df["generation"] = df["generation"].replace(" ", "")
df['generation'] = df['generation'].str.split('-').str[-1].map(romanos)

lista = []
for i in range(1, len(df)):
    if pd.isna(df["generation"].iloc[i]):
        # para o Pandas aceitar a alteração real no loop idêntico, usa-se o .iat
        df.iat[i, df.columns.get_loc("generation")] = df["generation"].iloc[i - 1]
    
    lista.append(df["generation"].iloc[i])

print(len(lista))
# lista feita para verificar se o loop acima funcionou corretamente
```
